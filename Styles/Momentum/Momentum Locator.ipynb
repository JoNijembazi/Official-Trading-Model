```python
# Create a list of dataframes for each stock
list_of_stocks = []

for basket in Data_list:
    for stocks in basket['Close'].columns:  # Go through each stock in the basket of names

        cols = pd.MultiIndex.from_tuples([(f'{stocks}', 'Date'), (f'{stocks}', 'Open'), (f'{stocks}', 'Close'), (f'{stocks}', 'High'), (f'{stocks}', 'Low'), (f'{stocks}', 'Volume')])
        
        data = [basket['Close'][stocks].index.values, 
                basket['Open'][stocks].values, basket['Close'][stocks].values, basket['High'][stocks].values, 
                basket['Low'][stocks].values, basket['Volume'][stocks].values]
        
        # Transpose the data to match the number of columns
        data = np.array(data).T
        ticker_data = pd.DataFrame(data, columns=cols)
        
        ticker_data[(f'{stocks}', 'Date')] = pd.to_datetime(ticker_data[(f'{stocks}', 'Date')], unit='ns')
        list_of_stocks.append(ticker_data)
```

```python
Final_stats = pd.DataFrame(columns=['Ticker', 'Mean Return', 'Return Standard Deviation', 'Mean High-Low Spread % ', 'High-Low Spread Standard Deviation %','Return/Spread Ratio','Open/Close Ratio'])
for stock in list_of_stocks:
    for col in stock.columns.levels[0]:
        stock[(col, 'Return')] = stock[(col, 'Close')].pct_change()
        stock[(col, 'High-Low Spread %')] = (stock[(col, 'High')] - stock[(col, 'Low')])/stock[(col, 'Open')]
        stock[(col, 'Open-Close Spread %')] = abs(stock[(col, 'Open')] - stock[(col, 'Close')])/stock[(col, 'Open')]
        Final_stats = pd.concat([Final_stats,
                            pd.DataFrame(data=[[f'{col}', stock[(col, 'Return')].mean()*100, stock[(col, 'Return')].std()*100, 
                                stock[(col, 'High-Low Spread %')].mean()*100, stock[(col, 'High-Low Spread %')].std()*100,
                                (stock[(col, 'Return')].mean()/stock[(col, 'High-Low Spread %')].mean()), (stock[(col, 'Return')].mean()/stock[(col, 'Open-Close Spread %')].mean())
                                ]], 
                            columns=Final_stats.columns)], 
                            ignore_index=True)
```

```python
px.scatter(Final_stats, x='Mean Return', y='Return Standard Deviation', title='Mean Return & Volatility of Stocks', labels={'Mean Return':'Mean Return (%)', 'Return Standard Deviation':'Standard Deviation (%)'},text='Ticker').show()
```

```python
Final_stats.sort_values(by='Return/Spread Ratio', ascending=False, inplace=True)
px.bar(Final_stats, x='Ticker', y='Return/Spread Ratio', title='Return/Spread Ratio of Stocks', labels={'Return/Spread Ratio':'Return/Spread Ratio (x)'}).show()
```
